```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#pair plot to visualise the attributes
sns.pairplot(data=df, hue= 'Label')
```

```python
training_set, test_set = train_test_split(df, test_size = 0.2, random_state = 1)
```

```python
#data processing
x = df.iloc[:, [0]].values #0 column of data
y = df.iloc[:, -1].values #last column of data
```

```python
x
```

```python
y
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,0] = le.fit_transform(x[:,0])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.2, random_state = 1)
print("Training split input- ", x_train.shape)
print("Testing split input- ", x_test.shape)
```

```python
#Initializing Support Vector Machine and fitting the training data
classifier = SVC(kernel='rbf', random_state = 1)
classifier.fit(x_train,y_train)
```

```python
#Predicting the classes for test set
y_pred = classifier.predict(x_test)
```

```python
#Attaching the predictions to test set for comparing
test_set["Predictions"] = y_pred
```

```python
y_pred
```

```python
#Calculating the accuracy of the predictions
cm = confusion_matrix(y_test,y_pred)
accuracy = float(cm.diagonal().sum())/len(y_test)
print("\nAccuracy Of SVM For The Given Dataset : ", accuracy)
```
